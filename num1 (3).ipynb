```python
import random
trials=10000
even_rolls_count=0
for _ in range(trials):
  roll=random.randint(1,6)
  if roll%2==0:
    even_rolls_count+=1
probability_even=even_rolls_count/trials
probability_even
```

```python
S = {1, 2, 3, 4, 5, 6}
A = {1, 2}
B = {1, 3, 5}
union_AB = A.union(B)
intersection_AB = A.intersection(B)
complement_A = S.difference(A)
complement_B = S.difference(B)
print("연습02 결과:", union_AB, intersection_AB, complement_A, complement_B)
```
